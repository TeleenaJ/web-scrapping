```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://meesho.com/earrings-women/pl/6a15h"
```

```python
url
```

```python
allow=requests.get(url)
```

```python
allow
```

```python
soup=BeautifulSoup(allow.text)
```

```python
soup
```

```python
item_name=soup.select(".cQhePS")
```

```python
item_name
```

```python
object_name=[]
```

```python
for i in item_name:
    object_name.append(i.text)
```

```python
object_name
```

```python
price=soup.select(".hiHdyy")
```

```python
price
```

```python
earring_rate=[]
```

```python
for j in price:
    earring_rate.append(j.text)
```

```python
earring_rate
```

```python
import pandas as pd
```

```python
meesho_data=pd.DataFrame(object_name,columns=["product_name"])
```

```python
meesho_data
```

```python
meesho_data["product_rate"]=earring_rate
```

```python
meesho_data
```

```python
rating
```
